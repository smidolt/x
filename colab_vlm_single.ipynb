```python
# Run VLM orchestrator (no OCR/blocks)
!python -m src.orchestrator_vlm \
  --input input/google.jpg \
  --output output_vlm \
  --vlm-model-reasoner Qwen/Qwen2-VL-2B-Instruct \
  --vlm-device auto \
  --vlm-max-tokens 256 \
  --vlm-temperature 0.2
```

```python
from pathlib import Path

out_json = Path('output_vlm/json/google.vlm.json')
val_json = Path('output_vlm/validation/google.vlm.validation.json')
print('VLM JSON exists:', out_json.exists())
print('Validation JSON exists:', val_json.exists())
if out_json.exists():
    print(out_json.read_text()[:1000])
if val_json.exists():
    print(val_json.read_text())
```
